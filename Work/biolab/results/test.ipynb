```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("result.csv")
data
```

```python
fig = px.histogram(data, x="average_molecular_weight", color="status")
fig.show()
```

```python
data_1 = data[(data['average_molecular_weight']<1200) & ((data['status'] == 'quantified') |(data['status'] == 'detected'))]
```

```python
fig = px.histogram(data_1, x="average_molecular_weight", title = 'quantified and detected', nbins=100)
fig.show()
```
